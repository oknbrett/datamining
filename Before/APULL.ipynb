```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks

# Extract relevant data
time = df['secondsElapsed']  # Time column
gravity_z = df['wristMotion_gravityZ']  # Gravity Z-axis column

# Find peaks (highest points) and troughs (lowest points)
peaks, _ = find_peaks(gravity_z, height=0)  # Adjust height if needed
troughs, _ = find_peaks(-gravity_z, height=0)  # Invert to find lowest points

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(time, gravity_z, label='Gravity Z', color='blue')

# Mark reps
plt.scatter(time[peaks], gravity_z[peaks], color='red', label='Top of Rep', marker='^')
plt.scatter(time[troughs], gravity_z[troughs], color='green', label='Bottom of Rep', marker='v')

# Labels and title
plt.xlabel("Time (seconds)")
plt.ylabel("Gravity (Z-axis)")
plt.title("Gravity Influence During Exercise")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks

# Extract relevant data
time = df['secondsElapsed']  # Time column
accel_z = df['wristMotion_accelerationZ']  # Acceleration Z-axis

# Find the highest and lowest acceleration points
lengthened, _ = find_peaks(-accel_z, height=0)  # Inverted to detect the lowest (lengthened)
shortened, _ = find_peaks(accel_z, height=0)  # Direct to detect the highest (shortened)

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(time, accel_z, label='Acceleration Z', color='blue')

# Mark positions
plt.scatter(time[shortened], accel_z[shortened], color='red', label='Shortened Position (Top)', marker='^')
plt.scatter(time[lengthened], accel_z[lengthened], color='green', label='Lengthened Position (Bottom)', marker='v')

# Labels and title
plt.xlabel("Time (seconds)")
plt.ylabel("Acceleration (Z-axis)")
plt.title("Acceleration Influence During Exercise (Lengthened vs. Shortened)")
plt.legend()
plt.show()
```
